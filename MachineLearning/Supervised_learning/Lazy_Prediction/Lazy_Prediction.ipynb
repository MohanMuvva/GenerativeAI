```python
# Importing necessary libraries
from lazypredict.Supervised import LazyClassifier
from sklearn.metrics import precision_score, recall_score, accuracy_score, classification_report
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

# Load the dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize LazyClassifier
clf = LazyClassifier(ignore_warnings=True, custom_metric=None)
models, _ = clf.fit(X_train, X_test, y_train, y_test)

# For each model in LazyClassifier, fit on training data and predict on test data separately
for model_name, model in clf.models.items():
    # Fit the model on the training data
    model.fit(X_train, y_train)

    # Predict on the test data
    y_pred = model.predict(X_test)

    # Print metrics for each model
    print(f"Model: {model_name}")
    print(f"Precision: {precision_score(y_test, y_pred)}")
    print(f"Recall: {recall_score(y_test, y_pred)}")
    print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    print("-" * 60)
```

```python
from sklearn.metrics import precision_score, recall_score, classification_report
from lazypredict.Supervised import LazyClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

# Load the dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize LazyClassifier
clf = LazyClassifier(ignore_warnings=True, custom_metric=None)
models, predictions = clf.fit(X_train, X_test, y_train, y_test)

# Get the best model name from the sorted DataFrame
best_model_name = models.index[0]

# Retrieve the actual model instance from LazyClassifier's `models` dictionary
best_model = clf.models[best_model_name]

# Train the best model on the training data
best_model.fit(X_train, y_train)

# Predict on the test data
predictions = best_model.predict(X_test)

# If predictions are probabilities, convert them to binary predictions
if hasattr(best_model, "predict_proba"):
    predictions_binary = (predictions > 0.5).astype(int)
else:
    predictions_binary = predictions

# Calculate precision and recall
precision = precision_score(y_test, predictions_binary)
recall = recall_score(y_test, predictions_binary)

print(f"Best Model: {best_model_name}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print("Classification Report:")
print(classification_report(y_test, predictions_binary))
```
