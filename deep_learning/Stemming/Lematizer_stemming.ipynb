```python
import matplotlib.pyplot as plt

from nltk.tokenize import word_tokenize
from collections import Counter
import os
import nltk
from nltk.corpus import stopwords # Import stopwords

# Ensure the 'punkt' package is downloaded
nltk.download('punkt')
nltk.download('stopwords') # Download stopwords

# Read the text from the uploaded file
file_path = list(uploaded.keys())[0]
with open(file_path, 'r') as file:
    text = file.read()

# Tokenization and normalization (lowercase)
normalized_tokens = [word.lower() for word in word_tokenize(text)]
word_counts = Counter(normalized_tokens)

# Filter out stopwords
filtered_words = [word for word in normalized_tokens if word not in stopwords.words('english')]

# Update word_counts to only include filtered words
word_counts = Counter(filtered_words)

# Find the most common words
# Call most_common on the word_counts (Counter object) instead of filtered_words (list)
most_common_words = word_counts.most_common(15)  # Adjust the number as needed
words, counts = zip(*most_common_words)

print("Most Repeated Words:")
for word, count in zip(words, counts):
    print(f"'{word}' appears {count} times.")

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(words, counts)
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Most Frequent Words")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
import requests
from bs4 import BeautifulSoup
import re
from wordcloud import WordCloud
from collections import Counter
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
def scrape_and_analyze(url):
    # Fetch the content from URL
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract text from paragraphs and headings
    text = ' '.join(p.text for p in soup.find_all(['p', 'h1', 'h2', 'h3', 'h4', 'h5', 'h6']))

    # Tokenize and count words
    words = re.findall(r'\w+', text.lower())
    word_counts = Counter(words)
    # Filter out stopwords
    filtered_words = [word for word in words if word not in stopwords.words('english')]

    #filtered words
    word_counts = Counter(filtered_words)


    # Identify the most common words
    most_common_words = word_counts.most_common(10)

    # Plotting
    words, counts = zip(*most_common_words)
    plt.figure(figsize=(10, 6))
    plt.bar(words, counts, color='blue')
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.title('Top 10 Most Repeated Words in the Article')
    plt.xticks(rotation=45)
    plt.show()

    # Generate a word cloud image
    wordcloud = WordCloud(width = 800, height = 400, background_color ='white').generate(text)

    # Display the generated image:
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")  # Do not show axes to keep it clean
    plt.show()

# URL of the webpage to analyze
url = "https://neptune.ai/blog/tokenization-in-nlp"
scrape_and_analyze(url)
```
